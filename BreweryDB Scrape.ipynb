```python
beer_df=pd.read_csv('beer_review_ave.csv')
beer_df['brewery_name_simple'] = beer_df['brewery_name'].str.replace('Brewing Company', "")
beer_df['brewery_name_simple'] = beer_df['brewery_name_simple'].str.replace('Brewing Co.', "")
beer_df['brewery_name_simple'] = beer_df['brewery_name_simple'].str.replace('Brewing', "")
beer_df['brewery_name_simple'] = beer_df['brewery_name_simple'].str.replace('Restaurant & Brewery', "")
beer_df
```

# Web Scrape of BreweryDB

```python
#Importing libs
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import requests

chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option('useAutomationExtension', False)
browser = webdriver.Chrome('chromedriver', options=chrome_options)
url='https://www.brewerydb.com'
browser.get(url)

search = beer_df['brewery_name_simple'][4] + " " +beer_df['beer_name'][4]
    
#Look for the class
search_box = browser.find_element_by_class_name('header-search').find_element_by_id("topSearchBox")


#Write what we be searched
search_box.send_keys(search)

#Submit the text
search_box.send_keys(Keys.RETURN)

#get html to scrape
html=requests.get(browser.current_url).text
soup = BeautifulSoup(html, 'html.parser')
#get first url
beer_url=soup.find_all('a')
#if (beer_df.loc[beer_df.index[4],'brewery_name_simple'] + ", " + beer_df.loc[beer_df.index[0],'beer_name']) in search[4]:
    possible_links = soup.find_all('a')
    for link in possible_links:
        if (beer_df.loc[beer_df.index[4],'beer_name']) in link.get_text():
            first_link = link.get('href')
            break
browser.get(url+first_link)

# html=requests.get(browser.current_url).text
# soup = BeautifulSoup(html, 'html.parser')
# beer_elements = soup.find_all('div', _class='col-xs-3 text-center')
# abv=beer_elements[0].text
# ibu=beer_elements[1].text
# og=beer_elements[2].text
# srm=beer_elements[3].text

# print(f'ABV: {abv}, IBU: {ibu}, OG: {og}, SRM: {srm}')

# hops =
# malt =
# yeast=
# adjuncts=
```
