```python
import pandas as pd
```

```python
beer_reviews = pd.read_csv('beer_reviews.csv')
beer_reviews
```

```python
beer_review_ave = beer_reviews.groupby(['beer_beerid','beer_name','brewery_name','beer_style']).mean()
beer_review_ave
```

```python
beer_review_ave = beer_review_ave.reset_index()
# beer_review_ave = beer_reviews.reset_index(inplace=True)
beer_review_ave
```

```python
beer_review_ave.to_csv('beer_review_ave.csv')
```
